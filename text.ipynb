```python
# as test.ipynb 
import random
```

```python
class nakaliLLm:
    def __init__(self) : 
        print("LLM created")
    
    def predict(self , prompt):
        response_list = [
            
            "delhi is the capital of india",
            "Ipl is the cricket League",
            "AI stands for artificial inteligence"
        ]
        return {"response" :random.choice(response_list)}
    
        
```

```python
class FakePromptTemplate:
    def __init__(self , template , input_variable ):
        self.template = template
        self.input_variable = input_variable
    def format(self , input_dict):
        return self.template.format(**input_dict)
```

```python
template = FakePromptTemplate(
    template="Write a {length} song about {topic}",
    input_variable=[ "length", 'topic']
)
```

```python
prompt = template.format({"length" : "long", "topic" : "india"})
```

```python
llm = nakaliLLm()
llm.predict(prompt)
```
